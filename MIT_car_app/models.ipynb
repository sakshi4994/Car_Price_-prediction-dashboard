```python
x = df[['insurance_validity','fuel_type', 'seats', 'kms_driven', 'ownsership', 'transmission']]
y = df['price(in lakhs)']
x.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
models = {
          'LinReg':LinearRegression(),
          'KNN-3':KNeighborsRegressor(n_neighbors = 3),
          'SVM-Poly':SVR(kernel='poly')
         }
for name,model in models.items():
    model.fit(x,y)
    sc= model.score(x,y)
    print(f"{name} score: {sc:.4f}")
```

```python
final_model = models['KNN-3']
```

```python
d1={'Comprehensive':0,'Third Party insurance':1,'Zero Dep':2,'Not Available':3,'Third Party':1}
d2 = {'Petrol':0,'Diesel':1,'CNG':2}
d3={'Manual':0,'Automatic':1}
d4={'First Owner':1,'Second Owner':2, 'Third Owner':3, 'Fifth Owner':5}
```

```python
insurance_validity = input('Insurance validity (Comprehensive/Third Party insurance/Zero Dep/Not Available): ')
fuel_type = input('Fuel Type (Petrol/Diesel/CNG): ')
seats = input('Number of Seats: ')
kms_driven = input('KMs Driven: ')
ownership = input('Ownership (First Owner/Second Owner/etc.): ')
transmission = input('Transmission Type (Manual/Automatic): ')
test_input = [[
    d1[insurance_validity],
    d2[fuel_type],
    int(seats),
    int(kms_driven),
    d4[ownership],
    d3[transmission]
]]
# Prediction
yp = final_model.predict(test_input)[0]
print(f'\nPredicted Car Price is: {yp:.2f} lakhs')
```
